<span style="color:blue"> Son Algoritmos voracez </span>

Son voracez porque la funcion de coste en si misma, no es una funcion convexa. SIno que tinen max y min locales.

Normalmente caen en esos minimos locales que no te garantizan que sea el minimo global de la funcion de costo. ESO ES UN PROBLEMA pero en principio no tienen solucion.

El problmea es que si dejamos extender el arbol de forma grande, que todos los nodos terminen siendo HOJAS(**PUROS**)(Todos los patrones de la misma clase en una hoja)

Esto genera **OverFiting**, porque cunado llegan los patrones de Tesst pueden caer en una prediccion errornea.
