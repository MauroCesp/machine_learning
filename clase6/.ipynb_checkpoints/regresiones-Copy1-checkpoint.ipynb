<h1><span style="color:blue">Regresion lineal </span></h1>


<span style="color:blue">1- Regresion lineal simple (regresión)</span>

Se trata de encontrar una relacion entre una variable independiente y otra dependiente

Cuando la independiente tiene alguna modificacion(incremento, decremento) la dependiente tambien de forma proporcional

A continuacion tenemos una gráfica  con una nuvbe de puntos y una serie de rectas. 

Los puntos son los datos del dataset

QUeremos representar la relacion de la variable dependeinte f(x) con la variable i que es la independeinte.


<span style="color:red">Cual es la recta optima que explica esa relacion entre variables?</span>

La recta roja minimiza la distancia entre los puntos (datos) y la recta.
La linea recta es el resultado de una **regresión lineal**


En matematica se encuentra la recta averiguando dos cosas: 
- El corte con el eje **Y**
- La **pendiente** con el eje **X**

Entonces representado como ecuacion es:  y = wx +b

Que significa que y es igual a la pendiente por el eje x, más el conrte con el eje Y

Entonces este procedimiento lo que busca es encontrar el valor optimo para **W**(pendeinte) como para **B**(intersección), a partor de todos los puntos de los que disponemos (cada dato dentro del dataset)

Esto es para acomodara la recta . como equilibrarla en el punto optimo

```python
from IPython.display import Image
Image(filename='img/1.PNG') 
```

<span style="color:blue">2- Regresion lineal multiple</span>

Aqui la variable dependeinte y la independeinte buscan explicar un **HYPERPLANO**, porque aqui ya no buscamos explicar una relacion en dos dimensiones sino en multiples 

En lugar de encontrar una sola pendiente **W** buscamos encontrar multiples **Ws** y encontrar la pendiente

Se necesita resolver por medio de la optimizacion:
    
-**Minimos cuadrados ordinales** (esto lo vimos en fundamentos)

```python
from IPython.display import Image
Image(filename='img/2.PNG') 
```

**MODELO PARAMETRICO**

Son una serie de asumciones que nos permite resolver el problema de Minimos cuadrados

Aqui necesitamos **assumir** varias cosas:

    - que no hay **Colinealidad** entre variables(relaciones fuertes entre variables de entrada). Es decir que son INDEPENDIENTES
    
    - En el momento en donde existe correlacion el modelo comienza a subfrir y no ser tan bueno.
    
**TRUCO** Si podemos aplicarle a la variable de salida un REESCALADO STD(guardando el modelo rescalado para aplicarlo a nuevas variables) ,para que la variuable de salida termine con una media y una desviacion tipica entre 1 y 0
entonces esa variable va a tener una distribucion normal , dentro de los posible.

    - Se asume que hay autocorrelacion(valor independientes con los valores de las otras instancias.)

### Como implementarlo en Python

```python
import numpy, random, time
from sklearn.model_selection import train_test_split

# Este es la libreria en python que permite la regresion lineal
from sklearn.linear_model import LinearRegression

import pandas as pd

from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import RobustScaler, MinMaxScaler, StandardScaler
```

```python
file = 'machine.data'
#data = 'data.csv'
col_names = ['vendor','Model','MYCT','MMIN','MMAX','CACH','CHMIN ', 'CHMAX','PRP','ERP']

data = pd.read_csv(file, names = col_names)

data.drop('vendor', axis=1, inplace=True)

data.drop('Model', axis=1, inplace=True)

data.head()
```

#### Primero

Divido los datos entre los inputs y el output

```python
X = data[data.columns[:-1]]
Y = data['ERP']
```

#### Segundo

Divido con la funcion **train_test_split**. Esto es más para juguetear con los datos , no para realizar experimentos sistematizados

Le paso todo el conjunto de datos

Ento me divide en 70% para entrenamiento y 30% para test

Vamos a continuar utilizando <span style="color:red"> **X_train** y **y_train** </span> en los modelos de validación

```python
seed = random.randint(0,10)

# Aqui genero 4 dataframes nuevos (2 train, 2 test) 
#Usamos el 30% de instancias para test y 70% para train
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state=seed)
```

**Terecero**

Reescalamos los datos de entrada

Generamos el modelo de escalado mediante los datos de entrenamiento.

Y después transformamos los datos de entranamiento (X_train) mediante **stdScaler**

```python
stdScaler = MinMaxScaler().fit(X_train)

X_train = stdScaler.transform(X_train)
```

Y ahora instanciamos un objeto de la clase **linearRegression** y medinate **fit** lo alimentamos con los datos de entrenamiento.

```python
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

**Finalmente**

Llamo al metodo **predict** con las instancias que han sido reservadas para test . Es decir, **X_test**

Primero escalo los datos de X_test, antes de llamar al predict

```python
X_test = stdScaler.transform(X_test)
y_pred = modelo.predict(X_test)
```

Y como en este tipo de Regression el mejor opcion es el **Error cuadratico** pues lo saco.



```python
rmse = (numpy.sqrt(mean_squared_error(y_test, y_pred)))
```

Este resultado me va a decir el error cometido con relacion a la valiable de salida. Dependiendo de los que se trate el error se calcula en euros, puntos, metros...

```python
print(rmse)
```

Finalmente tenemos la posibilidad de ver los **W** a cada una de las variables en la ecuacion . Esto sirve para ver como afecta W a cada una de ellas.

```python
from IPython.display import Image
Image(filename='img/3.PNG') 
```

```python
Aqui podemos ver las variables mas correladas con el resultado que queremos predecir y que están influenciando en la ecuacion. En este caso son la 7-2-3
```

```python
print(modelo.coef_)
```

<h3><span style="color:green">Transformar Regression logistica(clasificacion binaria) a Rgresion multiclase(más de dos clases a predecir)</span></h3>

<span style="color:red">OVA - OVR (One vs ALL)</span>

Tengo un problema multiclase de 3 clases.

- Genera 3 regresiones logisticas (binario) (Una por cada clase)

```python
from IPython.display import Image
Image(filename='img/5.PNG') 
```

Entonces en cada regresion una de las clases será la **positiva** y las demás negativas. Y así sucesivamente genera las rectas en la gráfica.

SI el modelo es bueno , cuando entre un figura nueva a predecir irá descartando las que no son positivas hasta encontrar la figura. Las lineas tienen que dividir bien las clases.

Es por eso que se llama uno contra el resto. Este es el procedimeinto por default en **Scaler**

<span style="color:red">OVO(One vs One)</span>

Genera tantos modelos como pares hayan.

El problema es que si hay muchas clase tndre que genear muchos modelos.

Es un sistema de votacion y ganara aquella clase que más haya sido votada.

**Scaler** implemente esto de manera automatica. 

En el momento en que scaler encuentra en el **FIT** que en el set hay mas de dos clases, automaticamente activa el ONEvsONE


### Como implementarlo en Python

```python
import numpy, random, time
from sklearn.model_selection import train_test_split

# Este es la libreria en python que permite la regresion lineal
from sklearn.linear_model import LogisticRegression

import pandas as pd

from sklearn.metrics import roc_auc_score, confusion_matrix, accuracy_score, balanced_accuracy_score

from sklearn.preprocessing import StandardScaler

import sklearn
```

Voy a utilizar este dataset de clasificacion binaria. Pero solo tengo que cambiar el dataset a IRIS(multiclase)

El resto del codigo queda exactametne igual porque el algoritmo detecta en el **FIT** si la variable independiente es multiclase o no.

```python
file = 'magic04.csv'
#data = 'data.csv'
col_names = ['fLenght','fWidth','fSize','fConc','fConc1','fAsym','fM3Long ', 'fM3Trans','fAlpha','fDist', 'class']

data = pd.read_csv(file, names = col_names)
#data = pd.read_csv(file)

data.tail()
```

#### Primero

Divido los datos entre los inputs y el output

```python
X = data[data.columns[:-1]]
Y = data['class']
```

#### Segundo

Divido con la funcion **train_test_split**. Esto es más para juguetear con los datos , no para realizar experimentos sistematizados

Le paso todo el conjunto de datos

Ento me divide en 70% para entrenamiento y 30% para test

Vamos a continuar utilizando <span style="color:red"> **X_train** y **y_train** </span> en los modelos de validación

```python
seed = random.randint(0,10)

# Aqui genero 4 dataframes nuevos (2 train, 2 test) 
#Usamos el 30% de instancias para test y 70% para train
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=seed)
```

**Terecero**

Reescalamos los datos de entrada

Generamos el modelo de escalado mediante los datos de entrenamiento.

Y después transformamos los datos de entranamiento (X_train) mediante **stdScaler**

```python
stdScaler = MinMaxScaler().fit(X_train)

X_train = stdScaler.transform(X_train)
X_test = stdScaler.transform(X_test)
```

Intancio un objeto de tipo Logistic regression

con el metodo ***FIT** le inyecto los datos de entrenamiento (Variables de pendeinete y la independiente)

```python
modelo = LogisticRegression()
modelo.fit(X_train, y_train)
```

Utilizo metodo **predict** , nos devuelve las predicciones para esas variables independeintes de los patrones reservados para test

```python
y_pred = modelo.predict(X_test)
```

Ya con esas predicciones averiguo las metricas

```python
cm = confusion_matrix(y_test, y_pred)
acc= accuracy_score(y_test,y_pred)

# Esto se utiliza para problemas desequilibrados
bacc = balanced_accuracy_score(y_test, y_pred)
```

```python
print(cm)
print(acc)
print(bacc)
print(sklearn.metrics.cohen_kappa_score(y_test,y_pred))
```
