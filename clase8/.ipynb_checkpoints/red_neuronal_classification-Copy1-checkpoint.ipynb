```python
from pandas import read_csv 
from sklearn.neural_network import MLPRegressor 
import time, random 
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error 

import matplotlib.pyplot as plt

seed=random.seed(time.time()) 
```

PAra el ejercicio utilizo un problema de clasificacion **BINARIA**

- G

- H

```python
seed=random.seed(time.time()) 

filename = 'comptagesvelo2015.csv'  
#data = read_csv(filename, names=col_names)  

data = read_csv(filename) 

data.head()
```

#### Primero

Divido los datos entre los inputs y el output

```python
X = data[data.columns[:-1]] 


# NEcesito encontrar la Y
Y = data['class']  
```

#### Segundo

Divido con la funcion **train_test_split**

Le paso todo el conjunto de datos

Ento me divide en 70% para entrenamiento y 30% para test

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state=seed, stratify=Y)   
```

<h3><span style="color:red">Escalado de datos</span></h3>

Separar las variables que sean de tipo numerico de las categoricas.
Y con **EstandardScaler** le aplico un escaldo a esas variables numericas.

#### **NUMERICAS** ---------------> StandardScaler

```python

scaler =StandardScaler()
X_train = scaler.fit_transform(X_train) 

X_test= scaler.transform(X_test)  
```

<span style="color:red">La funcin de activacin: </span>Por defecto, la clase MLPClassifier tiene configurada la funcin de activacin ReLU. El parmetro activation permite cambiar esta opcin. 
- identity
- logistic
- tahn
- relu

Nmero y tamao de las capas ocultas: Se realiza a travs del parmetro <span style="color:red">hidden_layer_sizes</span> que utilizando una tupla de enteros determina el nmero de capas ocultas (tamao de la tupla) y el nmero de nodos por capa de la red. 
- La capa de entrada siempre tendr el nmero de dimensiones del problema y la de salida el nmero de clases del problema. 

El algoritmo de entrenamiento: Por defecto, la clase MLPClassifier tiene configurada el algoritmo Adam, un mtodo de descenso de gradiente estocstico.

La tasa de entrenamiento: Controla el tamao del paso en la actualizacin de los pesos. El parmetro que lo establece es learning_rate_init de tipo doble y por defecto con un valor de 0.001. Puede ser constante o dinmica. 

SOLVER -------------> El algoritmo de optimizacion que busca los mejores pesos que se adapten a este conjunto dentrenamiento.

- lbfgs (Este es el que funciona mejor) Es capaz de encontrar solucion ms exacta, pero mucho mas constoso computacionalmente.
    - Utilizarlo cunado se tienen datasets pequeos, porque funciona mucho mas rapido y mejor.
- sgd
- adam (Descenso de gradiente Estocastico) Este tambien funciona muy bien para sataset grandes.

Estas son todo los parametro que necesitamos ajustar bien

Aqui de una vez le meto los patrones de entrenamiento y ya queda listo 

```python

model = MLPRegressor(hidden_layer_sizes=(100,), max_iter=10000, activation='logistic', n_iter_no_change=50, solver='lbfgs').fit(X_train, y_train) 

```

Ahora tomamos los datos de test y los comparamos con el modelo ya entrenamo

```python
y_pred = model.predict(X_test)  
```
