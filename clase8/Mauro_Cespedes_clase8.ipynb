<h1><span style="color:blue"> Realiza el preprocesamiento más adecuado para los datos. </span></h1>

#### Primero

Divido los datos entre los inputs y el output.

Como este data set ya vienen dividido no tengo que dividir la parte de test sino que importo el dataset de test.

Cada linea representa los bits de una imagen

Al final se trata de comparar si coniscide con el numero de **OUTPUT**

Los numeros van de 0-9

```python
X_train = df[df.columns[:-1]] 

X_test = df1[df.columns[:-1]] 

y_train = df['output'] 

y_test = df1['output'] 
```

<h3><span style="color:red">Escalado de datos</span></h3>

Separar las variables que sean de tipo numerico de las categoricas.
Y con **EstandardScaler** le aplico un escaldo a esas variables numericas.

#### **NUMERICAS** ---------------> StandardScaler

```python
scaler =StandardScaler()
X_train = scaler.fit_transform(X_train) 

X_test= scaler.transform(X_test)
```

<h1><span style="color:blue"> Entrena un modelo de redes neuronales para clasificación y evalúa su
rendimiento.
 </span></h1>

```python
model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, activation='logistic', n_iter_no_change=50, solver='lbfgs').fit(X_train, y_train) 

```

Ahora tomamos los datos de test y los comparamos con el modelo ya entrenado

```python
y_pred = model.predict(X_test)  
```

```python
bacc =  balanced_accuracy_score(y_pred, y_test) 

acc =  accuracy_score(y_pred, y_test) 
cm =  confusion_matrix(y_pred, y_test)  

print(bacc, acc) 

print(cm) 
```

<h1><span style="color:blue"> Trata de buscar la mejor combinación para los distintos parámetros
del modelo </span></h1>

Creo otro dataframe para almacenar la infomación de rendimiento.

```python
import pandas as pd

cols = ['Función de activación','Layers ocultos','Algoritmo de optimización', 'Accuracy','Balanced accuracy']

df_2 = pd.DataFrame(columns=cols)


activation = ['identity', 'logistic', 'relu','tanh']

hiddenLayers = [10,50,100,200]

solver = ['lbfgs','sgd','adam']


for a in activation:
    for b in hiddenLayers:
        for c in solver: 
            
            model = MLPClassifier(hidden_layer_sizes=(b,), max_iter=1000, activation= a, n_iter_no_change=50, solver= c).fit(X_train, y_train)
            y_pred = model.predict(X_test)
            
            bacc =  balanced_accuracy_score(y_pred, y_test) 

            acc =  accuracy_score(y_pred, y_test) 
            cm =  confusion_matrix(y_pred, y_test)
            
            
            new_row = {'Función de activación':a, 'Algoritmo de optimización':c, 'Layers ocultos':b, 'Accuracy':acc, 'Balanced accuracy':bacc}
            df_2 = df_2.append(new_row, ignore_index=True) 

df_2
```

Ahora lo que quiero es encontrar a partir de este nuevo dataset la combinacion que mejor y pero funciona, através de un FOR loop.

```python
for i in df_2.index:
    if (df_2['Accuracy'][i] == df_2['Accuracy'].max()) and (df_2['Balanced accuracy'][i] == df_2['Balanced accuracy'].max()): 

        print('La combinación que MEJOR funciona es :')
        print(df_2.iloc[[i]])
        print('---------------------------------------')
        
    else:
        pass
    
    if (df_2['Accuracy'][i] == df_2['Accuracy'].min()) and (df_2['Balanced accuracy'][i] == df_2['Balanced accuracy'].min()):
        
        print('La combinación que PEOR funciona es :')
        print(df_2.iloc[[i]])
        print('---------------------------------------')
        
    else:
        pass   
```
