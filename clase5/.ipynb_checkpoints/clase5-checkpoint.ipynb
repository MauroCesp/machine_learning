## Primera técnica: Train and Test

DIvidirlos conjuntos de datos en dos cuinjunto 

Esta es la base en la que se sirven otras tecnicas

**Entrenamiento y test (validacion)**

Pero como tenemos en los apuntes, el conjunto de test lo dejaremos unicamente para la **VALIDACION** del rendimiento predictivo de nuestro modelo

### Particionado de datos (train-test) con Scaler

```python
from pandas import read_csv 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import random
import pandas

data = read_csv('indian_liver_patient.csv')

#MIssing values
data.fillna(0,inplace=True)


data.head()
```

#### Primero

Divido los datos entre los inputs y el output

```python
X = data[data.columns[:-1]]
X = pandas.get_dummies(X, prefix=['gen'])
Y = data['Dataset']


```

#### Segundo

Divido con la funcion **train_test_split**

Le paso todo el conjunto de datos

Ento me divide en 70% para entrenamiento y 30% para test

```python
test_portion = 0.3

seed = random.randint(0,10)

# Aqui genero 4 dataframes nuevos (2 train, 2 test) 
#Usamos el 30% de instancias para test y 70% para train
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = test_portion, random_state=seed)
```

Con esto ya sería suficiente para entrenar un modelo y  validar la precision para un problema de clasificacion en este caso.

```python
model = KNeighborsClassifier(5)
model.fit(X_train, y_train)

acc = model.score(X_test, y_test)
print('Acc:', round(acc*100.0,4))
```

## Validacion cruzada

Es para que la estimacion se haga con patrones random independendientes.

Hay tres tipos:
    
    - K-fold
    - Leaveoneout
    - Monte Carlo

### K-FOLD

El conjunto de datos se divide en **K** particiones

Se divide de manera aleatoria

Cada **FOLD** queda dentro de una particion



```python
from IPython.display import Image
Image(filename='img/1.PNG') 
```

Es decir, cada patron es introducido solamente en un fold. Todos son del mismo tamaño

En este ejemplo tenemos un 4-fold en donde divido mi conjunto en 4 sets.

Cojo 3 de los folds para TRAINING y el otro para TEST

Pero también puedo ir generando diferentes rendimiento cambiendo el fold que utilizo para test.

```python
from IPython.display import Image
Image(filename='img/2.PNG') 
```

A esto le samos cun MAEDIA y una **DESVIACION TIPICA**

la **desviacion_tipica** es muy importante porque cuando samos varios rendimeintos les samos esto y nos dice que esta ocurriendo dentro del conjunto de datos.

Lo optimo seriá tener una desviacion tipica **pequeña** porque asi la media sería más representativa.

Todos los modelos han sido generados con distintos datos. Se ha sacado un rendimiento para cada caso



#### Como saber el valor de K?

Lo más aceptado es 5 0 hasta 10

Si tenemos un conjunto de datos suficientemente grande se podría utilizar un 3

Garantizamos que todos los patrones hayan sido utilizados para TEST en una ocasion.

**Implementacion de K-fold:**

```python
## Son los mismos imports de arriba más éste
from sklearn.model_selection import KFold

from pandas import read_csv 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import random
import pandas

data = read_csv('indian_liver_patient.csv')

#MIssing values
data.fillna(0,inplace=True)
```

```python
X = data[data.columns[:-1]]
X = pandas.get_dummies(X, prefix=['gen'])
Y = data['Dataset']
```

```python
# Definimos el número de particiones
k = 10
seed = random.randint(0,10)
```

Instancia un nuevo objeto de la clase KFold
**n_plit** indico el numero de particiones
**shuffle** es un proceso completamente aleatorio

```python
kfold = KFold(n_splits=k, random_state=seed, shuffle= True)
```

Una vez que tenga esto genero el modelo predictivo **KNN**

```python
model = KNeighborsClassifier()
```

Ahora aqui es lo importante donde utilizo la funcion **cross_var_score**

Lo que hace es, atraves del parametro **cv** (pasamos el objeto generado de tipo KFOLD)

En este caso va a particionarlo aleatoriamente en 10 partes 

Nos arroja un resultado de cada un ade las interacciones que haayamos generado.

```python
from sklearn.model_selection import cross_val_score

res = cross_val_score(model,X,Y,cv=kfold)

## Aqui muestro la MEDIA y la DESVIACION STD para registrar el rendimiento de estemodelo
print("Acc: %.3f+/.%.3f"% (res.mean(),res.std()))
```

#### Implementacion leave-ove-out

```python
## Son los mismos imports de arriba más éste
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score

from pandas import read_csv 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import random
import pandas

data = read_csv('indian_liver_patient.csv')

#MIssing values
data.fillna(0,inplace=True)
```

```python
X = data[data.columns[:-1]]
X = pandas.get_dummies(X, prefix=['gen'])
Y = data['Dataset']
```

```python
loocv = LeaveOneOut()
model = KNeighborsClassifier()

# Igual que antes lo unico es introducir el objeto loovp dentro del parametro CV (Cross validation)
results = cross_val_score(model, X,Y, cv=loocv)

## Aqui muestro la MEDIA y la DESVIACION STD para registrar el rendimiento de estemodelo
print("Acc: %.3f+/.%.3f"% (results.mean(),results.std()))
```

## Monte Carlo

Le das un numero máximo de interacciones y aleatoriamente ira seleccionando patrones para test y para train. 

EL número de patrones para cada partición tambien sera aleatorio.

No te garantiza que algun patron simpre haya estado en el conjunto de train o de test, como si lo garantizan las tecnicas anteriores.

Es una situacion intermedia entre los dos ateriores.


```python
## Son los mismos imports de arriba más éste
from sklearn import model_selection
from sklearn.model_selection import cross_val_score

from pandas import read_csv 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import random
import pandas

data = read_csv('indian_liver_patient.csv')

#MIssing values
data.fillna(0,inplace=True)
```

```python
X = data[data.columns[:-1]]
X = pandas.get_dummies(X, prefix=['gen'])
Y = data['Dataset']
```

```python
# Definimos el número de particiones
k = 100
seed = random.randint(0,10)
test_portion= 0.1
```

```python
montecarlo = model_selection.ShuffleSplit(n_splits=k, test_size=test_portion, random_state=seed)

model = KNeighborsClassifier()

# Igual que antes lo unico es introducir el objeto loovp dentro del parametro CV (Cross validation)
res = model_selection.cross_val_score(model,X,Y,cv=montecarlo)

## Aqui muestro la MEDIA y la DESVIACION STD para registrar el rendimiento de estemodelo
print("Acc: %.3f+/.%.3f"% (res.mean(),res.std()))
```
