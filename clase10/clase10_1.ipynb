<h1><span style="color:blue"> BAGGING </span></h1>


<span style="color:red"> 1 - MUESTREO ALEATORIO CON REPETICION  </span>

Es un modelo de emsemble que funciona en paralelo, formado por una serie de modelos base.

TEnemos un dataset de entrenamiento y hacemos un **MUESTREO ALEATORIO CON REPETICION**

Significa que selecciono patrones al azar, y con esoso realizamos un primer muestreo.
Con esso patrones base genero el modelo base 1.

Despues hago un segundo muestreo. El patron puede ser elegido de nuevo. COn esos patrones hago el modelo base 2.

Nuevamente selecciono otros patrones y genero un modelo base 3.

Elegimos el modelos que queremos utilizar para los entrenamiento, usualmente los arboles de decision fucionan muy bien para esto.

```python
from IPython.display import Image
Image(filename='img/5.PNG') 
```


<span style="color:red"> 2 - Genero modelo PARALELO  </span>

Entrenamos con estos patrones de los tres modelos.

Cunado entra un nuevo patron utilizamos un esquema paralelo:
    
    - Regresion ----------- PROMEDIO
    - CLasificacion ------- Votacion
 
A cada modelo se le pasa el patron a predecir.

<h1><span style="color:blue"> STACKING </span></h1>

Son modelos base que realizan una predicciones. En base de esas predicciones se genero otro modelo predictivo que combina de manera **lineal** o **no lineal** las predicciones de los distintos modelos base.

Por ejemplo si tenemos Arboles de desicion como modelos base, podemos combinar su salida en forma de una red neuronal, para que no todos pesen lo mismo

Se introduce en la prediccion la red neuronal para que le de diferente importancia a cada modelo base.

<h2><span style="color:blue"> Compromiso Bias-varianza </span></h2>

El **Bias** representa la complejidad del modelo.

La **varianza** representa la estabilidad en su rendimiento.

La grafica representa en **X** la complejidad y el **Y** el error cometido

Cunado aumentamos la complejidad del modelo vamos obteniendo una varianza mayor.

```python
from IPython.display import Image
Image(filename='img/10.PNG') 
```

La linea endonde se cruza la linea ascendente de la **varianza** y la linea descendente del **BIAS** es donde tenemos la complejidad optima de nuestro modelo.

Por ejemplo cuando aumentamos el nuemro de neuronas de una capa oculta de una red neuronas  estamos aumentando la complejidad.

**Como sabemos como podemos aumentar o dismiuri complejidad en un modelo sde Ensemble?**

    - Bagging resta complejidad (reduce el Overfitting)
    - Boosting aumenta complejidad (va refinando el error)

Entonces si tenemos un modelo que los resultados estan siendo muy malos por falta de complejidad metemos **Bagging** (estabilizamos su rendimiento), y en el caso contrario **Boosting** (refinamos el error)
