Un problema de **clasificacion** donde tenenmos que detecta que severidad tienen la mas detectada

```python
import numpy as np
import matplotlib
import pandas as pd
```

```python
file = 'mammographic_masses.data'
col_names = ['BI-RADS','Age','Shape','Margin','Density','Severity']

df = pd.read_csv(file, names = col_names)
df.head()
```

Necesitamos trabajar esos datos faltantes

No se puede quedar asi

### <span style="color:blue">**Como se implementa en python (Impute)**</span>

```python
import numpy as np
from sklearn.impute import SimpleImputer
import pandas as pd
```

**SimpleImputer** utiliza los datos de entrenamiento para gneerar un modelo de imputacion

```python
# Lo primero es crear el objeto
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Alimento el modelo de imputacion con los datos de entrenamiento
imputer.fit(clean_data)

# Aplica el modelo de imputacion recien creado sobre los datos que se le pasen
newdata = imputer.transform(clean_data)

clean_data.head()
```

### <span style="color:blue">**Otra forma de Imputar datos perdidos**</span>

```python
# ------REPLACE -------------------

# En caos de que tengamos que hacerlo de forma manual
data = clean_data.replace(to_replace=np.nan, value=1)

data.head()
```

```python
# ------INterpolate -------------------

# Cambiarlo poraquelllso valores que pertenezcan a una serie temporal
data.interpolate()
data.head()
```

<center> <h2><span style="color:red">MULTI-VARIANTE</span></h2> </center>

```python
from sklearn.impute import KNNImputer
```

```python
multi_data  = df.copy()
```

```python
# Primero creo un objeto Imputer
imputer = KNNImputer()


# Despues le paso el dataframe completo
imputer.fit(multi_data)
```

```python
# Lo que hace e sgenerar un nuevo dataframe sin los valores perdido utilizamos este metodo de KNN
# Utiliza la metrica **Euclidea**
new_data = imputer.transform(multi_data)

new_data
```
